# Визуализация Mathplotlib:
Сравнение различных библиотек для визуализации данных: Matplotlib, Seaborn и Plotly.

## **Matplotlib** - используют для отрисовки всевозможных видов графиков. Это незаменимая библиотека для любого аналитика данных. Помимо этого, Matplotlib лежит в основе других библиотек, например Seaborn, которая представляет высокоуровневый интерфейс над Matplotlib.

```python
# Визуализация с использованием MathPlotLib
# ЛИНЕЙНЫЙ ГРАФИК

import pandas as pd
from matplotlib import pyplot as plt

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        # Устанавливаем размер гарфика
        plt.figure(figsize=(10, 6))
        # Рисуем график согласно переданным переменным х и у
        plt.plot(self.x_data, self.y_data)
        # Название графика
        plt.title(self.title)
        # Наименование оси х
        plt.xlabel(self.x_label)
        # Наименование оси у
        plt.ylabel(self.y_label)
        # Добавляем сетку
        plt.grid(True)
        # Рисуем график
        plt.show()

# Запускаем код
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    # Ссылка на файл в проекте
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    # Описание столбцов файла
    columns = ['id','Country','Year','Series','Currency','Value']
    # загрузка массива данных
    rates = pd.read_csv(url, header = 1, names=columns)

# ЛИНЕЙНЫЙ ГРАФИК
    # Определяем выбор данных (фильтрум данные РФ)
    subset = rates[rates['Country'] == 'Russian Federation']

    # Строим график, передаем необходимые переменные
    graph = Graph(title="Курс Валют", x_label="Год", y_label="USD to RUB", x_data=subset['Year'], y_data=subset['Value'])

    # Выводим график
    graph.plot()
```

```python
# Визуализация с использованием MathPlotLib
# КРУГОВАЯ ДИАГРАММА
import pandas as pd
from matplotlib import pyplot as plt

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        # Устанавливаем размер гарфика
        plt.figure(figsize=(10, 6))
        # Рисуем график согласно переданным переменным х и у
        plt.pie(self.y_data, labels=self.x_data, autopct='%1.1f%%')
        # Название графика
        plt.title(self.title)
        # Рисуем график
        plt.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    # Ссылка на файл в проекте
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    # Описание столбцов файла
    columns = ['id','Country','Year','Series','Currency','Value']
    # загрузка массива данных
    rates = pd.read_csv(url, header = 1, names=columns)

    # Определяем выбор данных (фильтрум данные 2005 года)
    subset = rates[rates['Year'] == 2005]

    # Строим график, передаем необходимые переменные
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'], y_data=subset['Value'])

    # Выводим график
    graph.plot()
```

# **Рекомендации по выбору Matplotlib:**

* Библиотека подходит для высококачественных статических графиков и сложных визуализаций.
*   Она предлагает широкий спектр типов графиков и обширные возможности настройки всех элементов визуализации.
*   Matplotlib стоит выбрать, если нужен детальный контроль над графиками.
*   Библиотека надёжна и имеет большое сообщество пользователей и разработчиков, что упрощает поиск справки, руководств и примеров. При поиске в сообществе всегда найдется кто-то, кто уже решил проблему (или создал новую).
*   Matplotlib подойдёт, если критична производительность при работе с большими объёмами данных.
*   Библиотека может быть медленнее и менее эффективна при построении графиков больших наборов данных, особенно по сравнению с более оптимизированными библиотеками построения графиков.
