3 - **КРУГОВЫЕ ДИАГРАММЫ**\
*В основном используются для визуализации долей или процентов, которые составляют часть от целого.*\
Круговая диаграмма наглядно показывает, как каждая категория (сегмент) соотносится с другим. Хорошо подходят для распределения продаж по регионам, для анализа рыночных долей компаний в определенной отрасли, процентное распределение клиентов по возрастным группам или категориям товаров.

В данном проекте используем сумму валют, чтобы посмотреть процентное соотношение уровня национальной валюты.

1. построение круговой диаграммы с использованием библиотеки **Matplotlib**

```python
import pandas as pd
from matplotlib import pyplot as plt

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        plt.figure(figsize=(10, 6))
        plt.pie(self.y_data, labels=self.x_data, autopct='%1.1f%%')
        plt.title(self.title)
        # plt.grid(True)
        plt.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id','Country','Year','Series','Currency','Value']
    rates = pd.read_csv(url, header = 1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'], y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```

2. построение круговой диаграммы с использованием библиотеки c **Seaborn**

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        plt.figure(figsize=(10, 6))
        sns.set(style="whitegrid")
        plt.pie(self.y_data, labels=self.x_data, autopct='%1.1f%%', colors=sns.color_palette("Set2", len(self.y_data)))
        plt.title(self.title)
        plt.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id', 'Country', 'Year', 'Series', 'Currency', 'Value']
    rates = pd.read_csv(url, header=1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]
    # print(subset['Country'], subset['Value'])

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'],
                  y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```

3. построение круговой диаграммы с использованием библиотеки c **Plotly**

```python
import pandas as pd
import plotly.express as px

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        fig = px.pie(self.y_data, names=self.x_data, values=self.y_data, title=self.title)
        fig.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id','Country','Year','Series','Currency','Value']
    rates = pd.read_csv(url, header = 1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'], y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```
