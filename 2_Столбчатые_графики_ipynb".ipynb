**2 - СТОЛБЧАТЫЕ ГРАФИКИ (ДИАГРАММЫ)**\
*Они особенно полезны, когда нужно сравнить количество или величину разных категорий данных.*\
Хорошо подходят для сравнения данных, которые принадлежат разным категориям (сумма продаж за месяц, сравнение оценок учеников), для визуализации абсолютных величин (количество произведенных товаров на каждом заводе), для отображения части целого (доля отдела в общей прибыли).

1. построение столбчатой диаграммы с использованием библиотеки **Matplotlib**

```python
import pandas as pd
from matplotlib import pyplot as plt

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        plt.figure(figsize=(10, 6))
        plt.bar(self.x_data, self.y_data)
        plt.title(self.title)
        plt.xlabel(self.x_label)
        plt.ylabel(self.y_label)
        plt.grid(True)
        plt.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id','Country','Year','Series','Currency','Value']
    rates = pd.read_csv(url, header = 1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]
    # print(subset['Country'], subset['Value'])

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'], y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```

2. построение столбчатой диаграммы с использованием библиотеки **Seaborn**

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        plt.figure(figsize=(10, 6))
        sns.barplot(x=self.x_data, y=self.y_data)
        plt.title(self.title)
        plt.xlabel(self.x_label)
        plt.ylabel(self.y_label)
        plt.grid(True)
        plt.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id', 'Country', 'Year', 'Series', 'Currency', 'Value']
    rates = pd.read_csv(url, header=1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]
    # print(subset['Country'], subset['Value'])

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'],
                  y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```

3. построение столбчатой диаграммы с использованием библиотеки **Plotly**

```python
import pandas as pd
import plotly.express as px

class Graph:
    def __init__(self, title, x_label, y_label, x_data, y_data):
        self.title = title
        self.x_label = x_label
        self.y_label = y_label
        self.x_data = x_data
        self.y_data = y_data

    def plot(self):
        fig = px.bar(x=self.x_data, y=self.y_data)
        fig.update_layout(title=self.title, xaxis_title=self.x_label, yaxis_title=self.y_label)
        fig.show()

# Пример использования класса
if __name__ == "__main__":

    # Загружаем датасет Currency_Rates в массив pandas
    url = "https://raw.githubusercontent.com/irinawing/DIPLOMA/refs/heads/main/Exchange_Rates.csv"
    columns = ['id','Country','Year','Series','Currency','Value']
    rates = pd.read_csv(url, header = 1, names=columns)

    # Определяем выбор данных
    subset = rates[rates['Year'] == 2005]
    # print(subset['Country'], subset['Value'])

    # Построим график
    graph = Graph(title="Курс Валют 2005", x_label="Страна", y_label="USD to RUB", x_data=subset['Country'], y_data=subset['Value'])

    # Выводим график график
    graph.plot()
```
